```
def group_and_barplot(groupby, feat_agg='price_value', agg_func=[np.mean, np.median, np.size], feat_sort='size', top=50, subplots=True):
    return(
        df
        .groupby(groupby)[feat_agg]
        .agg(agg_func)
        .sort_values(by=feat_sort, ascending=False)
        .head(top)
    ).plot(kind='bar', subplots=subplots, figsize=(17, 5))
```

```
group_and_barplot('param_marka-pojazdu');
```

```
group_and_barplot('param_kraj-pochodzenia');
```

```
group_and_barplot('param_kolor', feat_sort='mean');
```
